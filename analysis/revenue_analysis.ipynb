```python
import pandas as pd
```

```python
orders = '../databases/orders.csv'
order_items = '../databases/order_items.csv'
users = '../databases/users.csv'
```

Read csv_files

```python
orders_df = pd.read_csv(orders)
order_items_df = pd.read_csv(order_items)
```

Convert created_at in orders table to datetime

```python
orders_df["created_at"] = pd.to_datetime(
    orders_df["created_at"],
    format="ISO8601"
)
```

```python
orders_df
```

Join orders_df and order_items_df

```python
merged_df = orders_df.merge(
    order_items_df,
    on="order_id",
    how="inner"
)
```

Where condition at status column in orders table

```python
filtered_df = merged_df[
    ~merged_df["status_x"].isin(["Cancelled", "Returned"])
]
```

Date trunc created_at column in orders table to month

```python
filtered_df["order_month"] = (
    filtered_df["created_at_x"]
    .dt.to_period("M")
    .dt.to_timestamp()
)
```

```python
filtered_df
```

Group joined tables by order_month and order_id column and sum the sale_price

```python
monthly_orders = (
    filtered_df
    .groupby(["order_month", "order_id"], as_index=False)
    ["sale_price"]
    .sum()
)
```

```python
monthly_orders
```

```python
monthly_orders = monthly_orders.rename(columns={
    "sale_price": "order_revenue"
})
```

```python
monthly_orders
```

```python
monthly_orders["order_revenue"].sum()
```

```python
monthly_summary_df = (
monthly_orders
.groupby("order_month", as_index=False)
.agg(
    num_orders=("order_id", "count"),
    gmv=("order_revenue", "sum"),
    aov=("order_revenue", "mean")))
monthly_summary_df
```
